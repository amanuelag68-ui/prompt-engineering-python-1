```python
!pip -q install google-genai

from google import genai
from google.colab import userdata

api_key = userdata.get("GEMINI_API_KEY")
if not api_key:
    raise ValueError("Missing GEMINI_API_KEY in Colab Secrets.")

client = genai.Client(api_key=api_key)

model = "gemini-flash-latest"

def call_llm(prompt):
    response = client.models.generate_content(
        model=model,
        contents=prompt
    )
    return response.text
```

```python
original_prompt = """
Summarize the following business idea in one paragraph:

Email Blaster is a SaaS platform that allows small businesses to send marketing emails.
It includes automation tools, analytics dashboards, audience segmentation,
and a freemium pricing model with paid upgrades.
"""

original_summary = call_llm(original_prompt)

print("ORIGINAL SUMMARY:\n")
print(original_summary)
```

```python
reflection_prompt = f"""
You previously wrote this summary:

{original_summary}

Now evaluate it using the following criteria:

1. Clarity (Is it specific and easy to understand?)
2. Audience fit (Is it written for business stakeholders?)
3. Completeness (Does it explain value proposition clearly?)
4. Conciseness (Maximum 120 words)
5. Professional tone

First:
- Briefly critique the summary against each criterion.

Then:
- Provide a revised improved version that meets ALL criteria.
- Keep it under 120 words.
- Make it more persuasive and structured.
"""

improved_output = call_llm(reflection_prompt)

print("SELF-REFLECTION + IMPROVED VERSION:\n")
print(improved_output)
```
